<h1>Amostragem de Sinais Senoidais</h1>

Considere que você pretende desenvolver um sistema embarcado para fazer a avaliação da qualidade da energia elétrica analisando o sinal da tensão da rede de alimentação com técnicas de processamento digital de sinais.

Sabe-se que o sinal da tensão da rede de alimentação é (aproximadamente) senoidal e possui frequência de 60 Hz.

Considerando esses fatos, qual é a melhor **Taxa de Amostragem** a ser escolhida para esta tarefa?

Considere que este sinal possa ser descrito como:

$$x(t) = cos(2 \pi \cdot 60 \cdot t)$$

e o processo de amostragem consistem em realizar a seguinte operação:

$$x[n] = x_c(n \cdot Ts)$$

Portanto, o sinal discretizado terá a seguinte expressão:

$$x[n] = cos(2 \pi \cdot 60 \cdot n \cdot Ts)$$


```python
import numpy as np
import matplotlib.pyplot as plt
```


**1)** Faça uma simulação de forma a obter as 50 primeiras amostras deste sinal discretizado a uma Taxa de Amostragem de **1200 Hz**.
Observe os resultados e faça suas considerações. Essa taxa é adequada?


```python
# cria um vetor de inteiros [0 1 2 ... N] com comprimento de 50
n = np.arange(0, 50)

# frequencia de amostragem
fs = 1200
Ts = 1 / fs

# calcula as 50 primeiras amostras do sinal discretizado
x = np.cos(2*np.pi*60*n*Ts)

# plota o resultado
plt.stem(n, x)
plt.title('50 amostras do sinal de 60Hz amostrado a 1200Hz')
plt.show()
```
