```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
```

```python
# cria um vetor de 300 valores aleatorios
x = np.random.random(300)
```

```python
# plota o vetor x
plt.plot(x)
plt.show()
```

```python
# abre um arquivo de audio e armazena os valores em x
[fs,x] = wavfile.read('pno1.wav')

# normaliza o vetor de entrada para manter no intervalo [-1 1]
x = np.array(x)
x = x / 32768.0

# mostra os primeiros 10 valores de x
print(x[:10])

# mostra a frequencia de amostragem utilizada na gravacao do arquivo
print(fs)
```

```python
# mostra o sinal de audio no tempo
plt.plot(x)
plt.show()
```

```python
# copia para y um trecho de 200 amostras de x
y = x[10000:10200]

# mostra o sinal y no tempo
plt.plot(y)
plt.show()
```

```python
# calcula a Transformada de Fourier de y
Y = np.fft.fft(y)

# mostra o módulo da transformada de y (espectro em frequencias de y)
plt.plot(abs(Y))
plt.show()
```
