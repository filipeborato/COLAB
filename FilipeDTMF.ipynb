<h1> Dual-tone multi-frequency signaling (DTMF) </h1>

O DTMF é um padrão de sinalização em telecomunicações que utiliza sinais compostos de dois sinais senoidais de frequências distintas para representar os 16 dígitos de um teclado telefônico convencional.

Neste padrão há dois conjuntos de frequências específicas: as frequências altas (1209 Hz, 1336 Hz, 1477 Hz e 1633 Hz) e as frequências baixas (697 Hz, 770 Hz, 852 Hz e 941 Hz).

Um sinal composto pela soma de um sinal de frequência alta com outro de frequência baixa representa um dígito específico.

Detalhes deste padrão podem ser vistos em:

https://en.wikipedia.org/wiki/Dual-tone_multi-frequency_signaling

https://www.etsi.org/deliver/etsi_es/201200_201299/20123502/01.01.01_60/es_20123502v010101p.pdf



O script abaixo mostra um exemplo de como gerar um sinal senoidal qualquer e salvá-lo em um arquivo do tipo wave. Execute-o e teste outras possibilidades.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io.wavfile as wav
```

```python
# frequencia de amostragem 8 kHz
fs = 8000
Ts = 1 / fs
```

```python
# cria um vetor de inteiros [0 1 2 ... N] com comprimento de 8000
n = np.arange(0,2000)

# gera um sinal senoidal de 110 Hz com 1 seg de duração e 0.7 de amplitude
x = 0.7 * np.sin(2*np.pi*110*n*Ts) + 0.7 * np.sin(2*np.pi*10*n*Ts)
```

```python
# plota o sinal x no tempo
plt.plot(n, x)
plt.title('Trecho de 1 seg de um sinal de 110Hz')
plt.show()
```

```python
# salva o vetor x em um arquivo do tipo wave
x = x * 32768.0
wav.write('110hz.wav', fs, np.int16(x))
```
